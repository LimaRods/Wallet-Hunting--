```python
import pandas as  pd
import requests
import numpy as np
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
dune_api = os.getenv('DUNE_API')
url = f"https://api.dune.com/api/v1/query/3319571/results?api_key={dune_api}"
response = requests.get(url)
df_vertex = pd.DataFrame(response.json())
data = df_vertex.iloc[-1,-1]
df = pd.DataFrame(data)
order = ['unique_trader', 'platform', 'trading_vol', 'vol_all_platform', 'interactions_count', 'last_interact_date', 'activity','balance_amount_usd']
df = df[order]
# download
df.to_excel("traders_vertex.xlsx", index= False)

df
```
